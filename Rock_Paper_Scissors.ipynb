# Simple Rock, Paper, Scissors game


```python
import random

CHOICES = ["rock", "paper", "scissors"]

def get_user_choice():
    """Ask the user for their choice and validate input."""
    while True:
        raw = input("Choose rock, paper or scissors (or 'q' to quit): ").strip().lower()
        if raw in ("q", "quit", "exit"):
            return None

        # accept shortcuts r, p, s
        if raw in ("r", "p", "s"):
            mapping = {"r":"rock", "p":"paper", "s":"scissors"}
            return mapping[raw]
        if raw in CHOICES:
            return raw
        print("Invalid choice. Type rock, paper, scissors, or r/p/s. Try again.")

def get_computer_choice():
    """Randomly select and return the computer's choice."""
    return random.choice(CHOICES)

def determine_winner(user, comp):
    """Return 'user', 'computer' or 'draw' depending on who wins."""
    if user == comp:
        return "draw"

    # user wins in these combinations:
    wins = {
        ("rock", "scissors"),
        ("paper", "rock"),
        ("scissors", "paper"),
    }
    if (user, comp) in wins:
        return "user"
    return "computer"

def main():
    print("Welcome to Rock, Paper, Scissors!")
    user_score = 0
    comp_score = 0
    rounds = 0

    while True:
        user_choice = get_user_choice()
        if user_choice is None:
            print("Thanks for playing! Final score â€” You: {}  Computer: {}".format(user_score, comp_score))
            break

        comp_choice = get_computer_choice()
        result = determine_winner(user_choice, comp_choice)
        rounds += 1

        print(f"\nYou chose: {user_choice}")
        print(f"Computer chose: {comp_choice}")

        if result == "draw":
            print("It's a draw!")
        elif result == "user":
            print("You win this round! ðŸŽ‰")
            user_score += 1
        else:
            print("Computer wins this round.")
            comp_score += 1

        print(f"Score after {rounds} round(s): You {user_score} â€” Computer {comp_score}\n")

        # Ask if user wants to play again
        again = input("Play again? (y/n): ").strip().lower()
        if again not in ("y", "yes"):
            print("Good game! Final score â€” You: {}  Computer: {}".format(user_score, comp_score))
            break

if __name__ == "__main__":
    main()
```
